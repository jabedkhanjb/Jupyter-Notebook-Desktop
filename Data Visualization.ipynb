# Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.figure import Figure

xpoint = np.array([1, 2, 3, 4, 5, 6, 7, 8])
ypoint = np.array([4, 2, 3, 7, 8, 10, 5, 6])

shape = plt.figure(figsize = (10, 8), facecolor = "g", edgecolor = "c", linewidth = 5)
myshape = shape.add_axes([1,1,1,1])

plt.title("Data Visualization", color = "k", fontsize = "20")
plt.xlabel("X-Axis", fontsize = 15, color = "white")
plt.ylabel("Y-Axis", fontsize = 15, color = "white")

plt.xticks(xpoint, labels = (["One", "Two", "Three", "Four", "Five", "Six", "Seven", "Eight"]))
plt.xlim(0, 10)
plt.ylim(0, 15)

myshape.plot(xpoint, ypoint, "o-c", ms = 15, mfc = "m", mec = "c")


plt.legend(["Linear Graph"])
plt.show()
```
