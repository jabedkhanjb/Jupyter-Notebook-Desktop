# Machine Learning - Scatter Plot

<i>A scatter plot is a diagram where each value in the data set is represented by a dot.</i>

Here, we will take two arrays and the first one represents the age of cars and the last array represents the speed of those car

```python
import matplotlib.pyplot as mpl

x = [4,5,2,9,5,5,3,1,1,7,2,4,5,6,3,8,2,1,7,4]
print(f"Number of arrays in {len(x)}")

y = [85,75,96,60,86,87,98,112,117,64,91,88,73,65,88,63,69,105,68,93]
print(f"Number of arrays in {len(y)}")

mpl.scatter(x,y)
mpl.show()
```
