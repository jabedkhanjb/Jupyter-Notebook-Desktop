## Histogram 
#### Matplotlib Package
<i> To visualize the data set we can draw a histogram with the data we collected. 
    We will use the Python module Matplotlib to draw a histogram.</i>

```python
import numpy as np
import matplotlib.pyplot as graph

demo = np.random.uniform(0.0, 5.0, 500)
print(demo)
graph.hist(demo, 5)
graph.show()
```

<h3><b><i>Create an array with 100000 random numbers, and display them using a histogram with 100 bars:</i></b></h3>

```python
import numpy as np
import matplotlib.pyplot as mygraph

mydemo = np.random.uniform(0.0, 5, 100000)
print(mydemo)
mygraph.hist(mydemo, 1000)
mygraph.show()
```
