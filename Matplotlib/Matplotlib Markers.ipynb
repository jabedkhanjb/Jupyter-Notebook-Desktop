```python
import matplotlib.pyplot as plt
import numpy as np

xpoint = np.array([0, 1, 4, 7, 8, 13, 15])
ypoint = np.array([16, 2, 5, 26, 1, 1, 20])

plt.plot(xpoint, ypoint, marker = "*", ms = "10")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoint = np.array([0, 2, 2, 4, 4, 6, 6, 10, 10, 0])

plt.plot(ypoint, "h-.g", ms = "10", mec = "c", mfc = "m") # ms for marker size, mec for marker edge color, mfc for marker face color
plt.show()
```
