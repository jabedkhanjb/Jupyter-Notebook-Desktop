```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.figure import Figure

p1 = np.array([10, 20, 30, 40, 50, 60, 70])
p2 = np.array([50, 12, 15, 35, 48, 95, 65])

p3 = np.array([2, 4, 8, 12, 15])
p4 = np.array([42, 15, 98, 164, 32])

figure1 = plt.figure(figsize = (6, 4), facecolor = "g")

ax1 = figure1.add_axes([0.1,0.8,0.8,0.8])
ax1.plot(p1,p2, "h-m", ms = 15, mec = "c")
ax1.legend(["Graph 1"])

ax2 = figure1.add_axes([1,0.8,0.8,0.8])
ax2.plot(p3,p4, "o-c", ms = 15, mec = "k")
ax2.legend(["Graph 2"])


# title
ax1.set_title("Linear Graph 1")
ax2.set_title("Linear Graph 2")

# xlabel
ax1.set_xlabel("X-Axis 1")
ax2.set_xlabel("X-Axis 2")

# ylabel
ax1.set_ylabel("Y-Axis 1")
ax2.set_ylabel("Y-Axis 2")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

p1 = np.array([10, 20, 30, 40, 50, 60, 70])
p2 = np.array([50, 12, 15, 35, 48, 95, 65])

plt.xticks(p1,labels = ["a","b","c","d","e","f","g"])
plt.bar(p1,p2, width = 5, color = "m")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

p1 = np.array([10, 20, 30, 40, 50, 60, 70])
p2 = np.array([50, 12, 15, 35, 48, 95, 65])

plt.xticks(p1,labels = ["a","b","c","d","e","f","g"])
plt.barh(p1,p2, height = 5, color = "m")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

ranx = random.sample(range(1, 100), 20)
rany = []

for i in range(20):
    value = random.randint(1, 100)
    rany.append(value)


plt.bar(ranx, rany, width = 2, color = "darkred")
plt.show()

plt.pie(ranx, labels = rany)
plt.show()

plt.scatter(ranx, rany, c = ranx, cmap = "viridis")
plt.colorbar()
plt.show()
```
