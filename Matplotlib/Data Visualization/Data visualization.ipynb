# Data Visualization Using Matplotlib

#### Linear Graph

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

# Data
xpoint = np.array([1, 7, 9, 5, 10, 12, 15])
ypoint = np.array([4, 12, 18, 19, 20, 10, 13])

myfigure = plt.figure(figsize = (10, 6), facecolor = "#9496a0", edgecolor = "c", linewidth = 3)
mydesign = myfigure.add_axes([1,1,1,1])
mydesign.plot(xpoint, ypoint, "h-c", ms = 15, mfc = "m", mec = "k")

plt.title("Data Visualization", fontsize = 20, color = "m")
plt.xlabel("X-Axis", fontsize = 15, color = "k")
plt.ylabel("Y-Axis", fontsize = 15, color = "k")

plt.xticks(xpoint, labels = ["One", "Two", "Three", "Four", "Five", "Six", "Seven"])

plt.xlim(0, 20)
plt.ylim(0, 25)


plt.legend(["Linear Graph"])
plt.show()


```
