```python
import matplotlib.pyplot as plt
import numpy as np

xpoint = np.array([10, 20, 30, 40])
ypoint = np.array([20, 25, 35, 55])

plt.plot(xpoint, ypoint, "h-y", ms=10, mec="c", mfc="m")
plt.title("Data Analysis", fontsize = 20, color="purple")
plt.xlabel("X-Axis", color="green", fontsize = 15)
plt.ylabel("Y-Axis", color="green", fontsize = 15)
plt.show()
```
